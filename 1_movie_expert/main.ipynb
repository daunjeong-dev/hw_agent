```python
import requests

BASE_URL = "https://nomad-movies-2.nomadcoders.workers.dev"
```

```python
def get_popular_movies():
    url = f"{BASE_URL}/movies"
    params = {
    }

    response = requests.get(url, params=params)
    response.raise_for_status()
    
    return response.json()
```

```python
results = get_popular_movies()
print(results[:3])
```

```python
def get_movie_details(movie_id):
    url = f"{BASE_URL}/movies/{movie_id}"
    params = {
    }

    response = requests.get(url, params=params)
    response.raise_for_status()

    return response.json()
```

```python
results = get_movie_details(550)
print(results)
```

```python
def get_movie_credits(movie_id):
    url = f"{BASE_URL}/movies/{movie_id}/credits"
    params = {
    }

    response = requests.get(url, params=params)
    response.raise_for_status()

    return response.json()
```

```python
results = get_movie_credits(550)
print(results[:3])
```

```python
FUNCTION_MAP = {
    "get_popular_movies": get_popular_movies,
    "get_movie_details": get_movie_details,
    "get_movie_credits": get_movie_credits,
}
```

```python
fcn_list = ['get_popular_movies()',
'get_movie_details("550")',
'get_movie_credits("550")']
```

```python
import re
```

```python
def run_fcn(function_call):
    match = re.match(r'(\w+)\("?(.*?)"?\)', function_call)

    if match:
        func_name = match.group(1)
        arg = match.group(2)
    
        if arg:
            result = FUNCTION_MAP[func_name](arg)
        else:
            result = FUNCTION_MAP[func_name]()
    
        return result
    
    return []
```

```python
run_fcn(fcn_list[1])
```

```python
for question in Que_list:
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": question}],
    )
    function_call = response.choices[0].message.content
    print(function_call)

    results = run_fcn(function_call)
    if isinstance(results, list):
        print(results[0])  # 첫 번째 요소만 출력
    elif isinstance(results, dict):
        print(results)     # 전체 출력
```
