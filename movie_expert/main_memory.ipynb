```python
import openai

client = openai.OpenAI()
```

```python
PROMPT = """
당신은 전문 영화 큐레이터입니다.

목표:
사용자와 자연스럽게 영화에 대한 대화를 나누며 사용자의 취향을 파악하고, 취향에 맞는 영화를 추천합니다.

행동 지침:
- 사용자의 영화 취향을 파악하기 위해 질문할 수 있습니다.
- 충분한 정보가 없다면 추가 질문을 통해 취향을 확인합니다.
- 사용자가 좋아하는 영화, 싫어하는 영화, 선호 장르, 배우, 감독, 분위기를 고려합니다.
- 사용자가 이미 본 영화는 가능한 한 추천하지 않습니다.
- 추천 시 추천 이유를 함께 설명합니다.
- 사용자의 피드백을 반영하여 추천을 개선합니다.

대화 스타일:
- 친근하고 자연스럽게 대화합니다.
- 영화에 대한 사용자의 의견을 존중합니다.
- 사용자의 취향에 맞는 작품을 우선 추천합니다.

"""
messages = []
messages.append({"role": "system", "content": PROMPT})
```

```python
def call_ai():
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
    )
    message = response.choices[0].message.content
    messages.append({"role": "assistant", "content": message})
    print(f"AI: {message}")
```

```python
while True:
    message = input("Send a message to the LLM...")
    if message == "quit" or message == "q":
        break
    else:
        messages.append({"role": "user", "content": message})
        print(f"User: {message}")
        call_ai()
```
